```python
import pandas as pd
tempcity = pd.read_csv('filtered_cities.csv')
tempcity.head()

```

```python
tempcity['AverageTemperature'] = pd.to_numeric(tempcity['AverageTemperature'], errors='coerce')
tempcity['AverageTemperatureUncertainty'] = pd.to_numeric(tempcity['AverageTemperatureUncertainty'], errors='coerce')
tempcity.dropna(subset=['AverageTemperature', 'AverageTemperatureUncertainty'], inplace=True)

tempcity['dt'] = pd.to_datetime(tempcity['dt'])
tempcity['Year'] = tempcity['dt'].dt.year

tempcity = tempcity.groupby(['Year', 'City', 'Country', 'Latitude', 'Longitude', 'NumericLatitude', 
                             'NumericLongitude'])[['AverageTemperature', 'AverageTemperatureUncertainty']].mean().reset_index()

tempcity.head()
```

```python
tempcity = tempcity[['Year', 'City', 'Country', 'AverageTemperature', 'AverageTemperatureUncertainty', 'Latitude', 'Longitude', 
                     'NumericLatitude', 'NumericLongitude']]
# Renomeie as colunas
tempcity = tempcity.rename(columns={
    'Year': 'Ano',
    'City': 'Cidade',
    'Country': 'País',
    'NumericLatitude': 'Latitude Numérica',
    'NumericLongitude': 'Longitude Numérica',
    'AverageTemperature': 'Temperatura Média',
    'AverageTemperatureUncertainty': 'Incerteza Temperatura Média'
})
tempcity = tempcity.drop(['Latitude', 'Longitude'], axis=1)
tempcity.head()
```
