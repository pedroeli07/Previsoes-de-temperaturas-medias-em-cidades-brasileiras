```python
import pandas as pd
temp = pd.read_csv('tabela_filtrada_organizada.csv')
display(temp)
```

```python
temp.isnull().sum()
temp.duplicated().sum()
```

```python
cidades_procuradas = ["Belo Horizonte", "Vitória", "São Paulo"]
cidades_presentes = temp[temp['Cidade'].isin(cidades_procuradas)]['Cidade'].unique()

if set(cidades_procuradas) == set(cidades_presentes):
    print("Todas as cidades estão presentes no DataFrame.")
else:
    cidades_ausentes = list(set(cidades_procuradas) - set(cidades_presentes))
    print("As seguintes cidades estão ausentes no DataFrame:", cidades_ausentes)
```

```python
cidades_de_interesse = ['Belo Horizonte', 'Vitória', 'São Paulo']

coordenadas = []

for cidade in cidades_de_interesse:
    filtro = temp['Cidade'] == cidade
    latitudes = temp.loc[filtro, 'Latitude Numérica'].unique()
    longitudes = temp.loc[filtro, 'Longitude Numérica'].unique()

    for lat, long in zip(latitudes, longitudes):
        coordenadas.append({'Cidade': cidade, 'Latitude': lat, 'Longitude': long})

coordenadas_df = pd.DataFrame(coordenadas)
print(coordenadas_df)
```

```python
# Crie um dicionário com as coordenadas das cidades de Belo Horizonte, Vitória e São Paulo
cidades_coordenadas = {
    'Belo Horizonte': (-20.09, -44.36),
    'Vitória': (-20.09, -40.95),
    'São Paulo': (-23.31, -46.31)
}

# Crie uma nova coluna chamada 'Nova Cidade' com base nas coordenadas
temp['Nova Cidade'] = temp.apply(lambda row: 
    next((cidade for cidade, coordenadas in cidades_coordenadas.items()
          if (row['Latitude Numérica'], row['Longitude Numérica']) == coordenadas), 'Outra cidade'), axis=1)

# Remova a coluna 'Cidade'
temp = temp.drop(columns=['Cidade'])
temp = temp.drop(columns=['País'])

# Renomeie a coluna 'Nova Cidade' para 'Cidade'
temp = temp.rename(columns={'Nova Cidade': 'Cidade'})

temp.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Suponha que você deseja escalar as colunas 'Latitude Numérica' e 'Longitude Numérica'
columns_to_scale = ['Latitude Numérica', 'Longitude Numérica', 'Incerteza Temperatura Média']

# Normalização (escala min-max)
#min_max_scaler = MinMaxScaler()
#temp[columns_to_scale] = min_max_scaler.fit_transform(temp[columns_to_scale])

# Padronização (z-score)
z_score_scaler = StandardScaler()
temp[columns_to_scale] = z_score_scaler.fit_transform(temp[columns_to_scale])
display(temp)
```

```python
cidades_procuradas = ["Belo Horizonte", "Vitória", "São Paulo", "Outra cidade"]
cidades_presentes = temp[temp['Cidade'].isin(cidades_procuradas)]['Cidade'].unique()

if set(cidades_procuradas) == set(cidades_presentes):
    print("Todas as cidades estão presentes no DataFrame.")
else:
    cidades_ausentes = list(set(cidades_procuradas) - set(cidades_presentes))
    print("As seguintes cidades estão ausentes no DataFrame:", cidades_ausentes)
```

```python
# Aplicar codificação one-hot na coluna 'Cidade'
temp = pd.get_dummies(temp, columns=['Cidade'], prefix='Cidade', drop_first=True)
display(temp)
```

```python
temp['Cidade_Outra cidade'] = temp['Cidade_Outra cidade'].astype(int)
temp['Cidade_São Paulo'] = temp['Cidade_São Paulo'].astype(int)
temp['Cidade_Vitória'] = temp['Cidade_Vitória'].astype(int)
temp.head()
```

```python
# Selecione as colunas relevantes
cols = ["Cidade_Outra cidade", "Cidade_São Paulo", "Cidade_Vitória"]

# Crie a coluna "Cidade_Belo Horizonte" com base na lógica desejada
temp["Cidade_Belo Horizonte"] = 1 - temp[cols].max(axis=1)

# Certifique-se de que a nova coluna seja do tipo inteiro
temp["Cidade_Belo Horizonte"] = temp["Cidade_Belo Horizonte"].astype(int)
display(temp)
```
